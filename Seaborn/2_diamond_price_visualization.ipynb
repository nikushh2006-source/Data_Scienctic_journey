```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

A Data frame with 53,940 rows and 10 variables

1.carat : Weight of the diamond (0.2 - 5.01)
2.cut : Quality of the cut (fair ,Good ,Premium ,Ideal)
3.color : Diamond color ,from D(best) to J (worst)
4.clarity : A measurement of how clear the diamond
5.depth : Total depth percentage = z / mean (x,y) =2*z/(x+y)(43--95)
6.table: Width of top diamond relative to widest point (43--95)
7.Price : price in US dollars($326 -- $18,823)
8.x =length in mm(0--10.74)
9.y=Width in mm(0--58.9)
10.z =Depth in mm(0--31.8)

Customizing color palettes 

```python
# Using a predefined palettes
figure =plt.figure(figsize=(12,3))
sns.set_context("notebook")
sns.set_palette("deep") #option deep muted ,bright ,pastel,dark,colorblind
sns.violinplot(x ="color" ,y ="price" ,data=df,hue='cut')
plt.title("violinplot with color palette")
plt.show()
```

```python
# creating  and using custom color palette
custom_palette =sns.color_palette(['#FF5733','#FF5667','#FF5643','#FF4325','#FF8768'])
sns.set_palette(custom_palette)
sns.countplot(x ='cut',data=df ,edgecolor ='black', hue='color')
plt.title('Count plot with Custom palette')
plt.show()
```

```python
# remove spines from the plot for a cleaner look
sns.set_style('white')
sns.histplot(df['price'] ,bins=10, edgecolor ='black')
sns.despine() # Remove the top right spines
plt.title("Histogram with spines remove")
plt.show()
```

```python
# Annotate point 
sns.scatterplot(x ='carat' ,y ='price' ,data=df)
plt.title("Scatter plot with Annotation")
for i in range(len(df)):
    plt.annotate(f"{df['price'][i]} " ,(df['carat'][i] ,df['price']) ,fontsize =8 ,alpha =0.7)
plt.show()    
```

```python
# Custmizing the legend
figure =plt.figure(figsize=(12,3))
sns.boxplot(x ='cut' ,y ='price' ,data=df)
plt.legend(title='Cut Types', labels=df['cut'].unique(), loc='upper left', bbox_to_anchor=(1, 1), frameon=True)
plt.title("Box plot with custom legend")

plt.show()
```
