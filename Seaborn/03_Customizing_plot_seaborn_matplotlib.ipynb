```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns

# Sample data
# load the sample diamond  data
df = pd.read_csv("Diamonds Prices2022.csv")
print(f"Number of rows before Sample : {df.shape[0]}")
print(f"Number of colunms before Sample : {df.shape[1]}")
print(f"{50*'-'}")
df =df.sample(5000)
print(f"Number of rows after Sample : {df.shape[0]}")
print(f"Number of colunms after Sample : {df.shape[1]}")


# Display the first few rows of dataset
df.head()


```

```python
# Set the title
sns.scatterplot(x ='carat' ,y ='price' ,data=df)
plt.title("Scatter plot of Carat VS Price")
plt.show()
```

```python
# Axes labeling
sns.scatterplot(x ='carat',y ='price' ,data=df)
plt.xlabel('Carat of Diamond')
plt.ylabel("Price of Diamond")
plt.show()
```

```python
# Customizing the legend
sns.scatterplot(x ='carat',y ='price' ,hue='cut',data=df)
plt.legend(title='Diamonds cut', loc='upper right')
plt.show()
```

```python
# Customizing ticks
sns.histplot(df['price'] ,bins=10 ,edgecolor ='black')
plt.xticks(rotation =45 ,fontsize =10)
plt.yticks(fontsize =10)
plt.show()
```

```python
# Adding the grid
sns.set_style("whitegrid")
sns.scatterplot(x ='carat' ,y ='price' ,data=df)
plt.grid(True ,linestyle ='--' ,alpha =0.5)
plt.show()
```

```python
# boxplot
sns.set_theme(style='whitegrid')
sns.boxplot(x ='cut' ,y = 'price' ,data=df ,palette='Set3')
sns.despine() #remove the top  and right spines for a clear look
plt.show()
```

```python
# figure 
figure =plt.figure(figsize=(10,6))#set figure size
sns.barplot(x ='cut' ,y ='price' ,data=df)
plt.title('Bar Plot of Price by cut (Large figure)')
plt.show()
```

```python
# Creating subplot 
figure,axes =plt.subplots(1,2 ,figsize =(12,3))
sns.histplot(df['carat'] ,bins=10 ,edgecolor ='black' ,ax =axes[0] ,kde=True)
axes[0].set_title('Histogram of Carat')
sns.histplot(df['price'] ,bins=10 ,edgecolor ='black' ,ax =axes[1] ,kde=True)
axes[0].set_title('Histogram of Price')
plt.show()
```

```python
# Adding Anotation
sns.scatterplot(x='carat',y='price',data=df)
plt.title("Scatter plot with Anotation")
for i in range(len(df)):
    plt.annotate(f" {df.iloc[i]['table']}", xy=(df.iloc[i]['carat'], df.iloc[i]['price']), fontsize=8)
plt.show()    
```

```python
# Customizing color palate
figure =plt.figure(figsize=(15,6))
sns.set_palette('muted')
sns.boxplot(x ='cut' ,y ='price' ,data=df ,hue='clarity')
plt.title("Boxplote with muted palatte")
plt.show()

```
