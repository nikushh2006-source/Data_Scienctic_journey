```python

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
# remove worning 
import warnings
warnings.filterwarnings('ignore')
```

```python
df =pd.read_csv("Diamonds Prices2022.csv")
print(df.shape)
df = df.sample(frac=0.1, random_state=10)
print(df.shape)
```

```python
df.head()
```

# Express plotly

## marginal plot useing plotly

```python
# Scatter plot with marginal histogram
fig = px.scatter(df,x ='carat' ,y ='price',
                 marginal_x='histogram',
                 marginal_y='histogram',
                 title='Marginal histogram with carat and price')
fig.show()
```

```python
# Scatter plot with marginal histogram
fig = px.scatter(df,x ='carat' ,y ='price',
                 marginal_x='box',
                 marginal_y='box',
                 title='Marginal histogram with carat and price')
fig.show()
```

```python
# Scatter plot with marginal histogram
fig = px.scatter(df,x ='carat' ,y ='price',
                 marginal_x='violin',
                 marginal_y='violin',
                 title='Marginal histogram with carat and price')
fig.show()
```

```python
# Scatter plot with marginal histogram
fig = px.scatter(df,x ='carat' ,y ='price',
                 marginal_x='histogram',
                 marginal_y='violin',
                 color='cut',
                 title='Marginal histogram with carat and price')
fig.show()
```

```python
# Scatter plot with marginal histogram
fig = px.scatter(df,x ='carat' ,y ='price',
                 marginal_x='histogram',
                 marginal_y='box',
                 color='cut',
                 facet_col='clarity',
                 facet_row='color',
                 title='Marginal histogram with carat and price')
# update figure size
fig.update_layout(height =1200 ,width =800)
fig.show()
```

# Animated Video plot 

```python
# Scatter plot 
fig =px.scatter(df ,x ='carat' ,y ='price' ,
                color='cut',
                size='price',
                animation_frame='cut',
                title='Diamond price vs  carat weighted ')
fig.show()
```

```python
# Scatter plot 
fig =px.scatter(df ,x ='carat' ,y ='price' ,
                color='cut',
                size='price',
                facet_col='color',
                # facet_row='clarity',
                animation_frame='cut',
                title='Diamond price vs  carat weighted ')
# remove legend
fig.update_layout(showlegend =False)
# update figure size 
fig.update_layout(height =400 ,width =1000)
# save in html
fig.write_html('Scatter_animated_plot.html')
fig.show()
```
