```python
# plt.pie( values ,labels=label_list,color= 'color_name',autopct ='%1.1f%%')  # pie chart use for percentage comparison
import matplotlib.pyplot as plt

regions =['North','South','East','West']
revenues = [1000,1500,800,1200]
plt.pie(revenues,labels=regions,colors=['red','yellow','green','orange'],autopct='%1.1f%%')
plt.title('Revenues Distribution by Region')
plt.show()
```
