```python
# plt.plot(x,y,color='color_name',linewidth=value,linestyle=line_style,marker=marker_symbol,label='label_name')

import matplotlib.pyplot as plt
months =[1,2,3,4]
sales = [1000,1500,1200,1800]
plt.plot(months,sales ,color = 'green',linestyle = '--',linewidth =2,marker ='o',label ='2025 Sales Data') 
plt.xlabel('Months')
plt.ylabel('Sales for month')
plt.title('Monthly Sales Data for 2025')
plt.legend(loc= 'lower right',fontsize =12)# loc ='upper left'
plt.grid(color ='gray',linestyle =':',linewidth =1)
plt.xlim(0,4)
plt.ylim(500,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```
