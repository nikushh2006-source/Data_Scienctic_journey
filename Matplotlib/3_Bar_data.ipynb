```python
# plt.bar( x_values , y_values , color='color_name', label='label_name')
import matplotlib.pyplot as plt
products =['A','B','C','D']
sales = [1000,1500,800,1200]
plt.bar(products,sales,color = 'orange',label='Sales Data')
plt.xlabel('Products')
plt.ylabel('Sales Amount')
plt.title('Sales Data Comparison')
plt.legend()
plt.show()# bar chart use for categorical data comparison
```
