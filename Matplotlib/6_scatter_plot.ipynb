```python
#plt.scatter(x_data,y_data,color='color_name',marker='marker_style',label='label_name')
import matplotlib.pyplot as plt

hours_studied =[1,2,3,4,5,6,7,8]
exam_scores = [35,40,50,55,65,70,75,80]

plt.scatter(hours_studied,exam_scores ,color ='green',marker='^',label ='Score of student')

plt.title('Hours studied vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.legend()
plt.grid(color ='gray',linestyle ='--',linewidth =0.5)
plt.show()
```
