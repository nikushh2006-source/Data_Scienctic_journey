```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

### these dataset contain  Borderlands of positive ,negative,neutral data

```python
df = pd.read_csv('twitter_training.csv')
df.head()
```

### Basic Data Cleaning

```python
df['Sentiment'] = df['Sentiment'].str.strip()
df.drop_duplicates(subset='Text', inplace=True)
df.dropna(inplace=True)
```

### Sentiment distribution 

```python
df['Sentiment'].value_counts().plot(kind='bar' ,figsize=(7,4),color =['blue' ,'orange' ,'green' ,'red'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation =45)
plt.show()
```

### Sentiment Count

```python
fig = plt.figure(figsize =(7,4))
sns.countplot(x=df['Sentiment'] ,hue='Sentiment' ,data=df)
plt.title('Sentiment Count')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

### Topic vise sentiment Heatmap

```python
pivot =pd.crosstab(df['Topic'] ,df['Sentiment'])

plt.figure(figsize=(6,4))
sns.heatmap(pivot ,
            annot=True,
            cmap='YlGnBu' )
plt.title('Topic vs Sentiment Heatmap')
plt.xlabel('Sentiment')
plt.ylabel('Topic')
plt.show()
```

### Sentiment Distribution

```python
fig =px.pie(df,names='Sentiment' ,title='Sentiment pie chart' ,hole=0.4)
fig.show()
```

### Tweet Length Analysis

```python
df['Tweet_Length'] = df['Text'].apply(len)

sns.histplot(df['Tweet_Length'], kde=True)
plt.title("Tweet Length Distribution")
plt.xlabel('Tweet length')
plt.ylabel('Frequency')
plt.show()
```
